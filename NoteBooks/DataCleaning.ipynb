```python
# | **Column**        | **Type**    | **Best Filling Strategy**                                           | **Why**                                                                         |
# | ----------------- | ----------- | ------------------------------------------------------------------- | ------------------------------------------------------------------------------- |
# | `price`           | Numeric     | ❌ **Do NOT fill with 0**<br>✅ Use **median or drop rows**           | Price = 0 ruins model. If target = price, drop rows with missing price.         |
# | `qty_ordered`     | Numeric     | ✅ Use **0 or median**                                               | 0 can mean no order. Median is better if you suspect data is missing, not zero. |
# | `grand_total`     | Numeric     | ✅ Recalculate: `price * qty_ordered`                                | Derived field — best to **compute it** rather than fill blindly.                |
# | `category_name_1` | Categorical | ✅ Use `"Unknown"` or `"Missing"`                                    | For ML models, placeholder categories work well.                                |
# | `discount_amount` | Numeric     | ✅ Fill with 0                                                       | Most likely means **no discount** was applied.                                  |
# | `payment_method`  | Categorical | ✅ Fill with `"Unknown"` or `"Missing"`                              | Keeps the row usable for classification or analysis.                            |
# | `Year`            | Numeric     | ✅ Fill with value extracted from `created_at` column (if available) | Reconstruct if possible. Else fill with 0 or drop if date matters.              |
# | `Month`           | Numeric     | ✅ Same as `Year`                                                    | Try to extract from any available datetime column.                              |
```

```python
# Drop rows where price is missing (important for prediction)
data = data[~data['price'].isnull()]

# Fill other numerical columns
data['qty_ordered'].fillna(0, inplace=True)
data['discount_amount'].fillna(0, inplace=True)



# Fill categorical
data['category_name_1'].fillna('Unknown', inplace=True)
data['payment_method'].fillna('Unknown', inplace=True)

# Fill Year and Month with 0 or infer from datetime column if available
data['Year'].fillna(0, inplace=True)
data['Month'].fillna(0, inplace=True)
```
